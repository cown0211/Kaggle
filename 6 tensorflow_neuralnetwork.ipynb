```python
x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype = np.float32)
y_data = np.array([[0], [1], [1], [0]], dtype = np.float32)

X = tf.placeholder(tf.float32, [None, 2])
Y = tf.placeholder(tf.float32, [None, 1])

W1 = tf.Variable(tf.random_normal([2, 5]), name = 'weight1')
b1 = tf.Variable(tf.random_normal([5]), name = 'bias1')
M1 = tf.sigmoid(tf.matmul(X, W1) + b1)

W2 = tf.Variable(tf.random_normal([5, 5]), name = 'weight2')
b2 = tf.Variable(tf.random_normal([5]), name = 'bias2')
M2 = tf.sigmoid(tf.matmul(M1, W2) + b2)

W3 = tf.Variable(tf.random_normal([5, 5]), name = 'weight3')
b3 = tf.Variable(tf.random_normal([5]), name = 'bias3')
M3 = tf.sigmoid(tf.matmul(M2, W3) + b3)

W4 = tf.Variable(tf.random_normal([5, 1]), name = 'weight4')
b4 = tf.Variable(tf.random_normal([1]), name = 'bias4')
y_hat= tf.sigmoid(tf.matmul(M3, W4) + b4)

# cost function
cost = -tf.reduce_mean(Y * tf.log(y_hat) + (1 - Y) * tf.log(1 - y_hat))
train = tf.train.GradientDescentOptimizer(learning_rate = 0.1).minimize(cost)

# Accuracy 
predicted = tf.cast(y_hat > 0.5, dtype = tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype = tf.float32))

# Launch graph
with tf.Session() as sess:
    # Initialzation
    sess.run(tf.global_variables_initializer())

    for step in range(10001):
        _, cost_val = sess.run([train, cost], feed_dict = {X: x_data, Y: y_data})
        if step % 100 == 0:
            print("Step: ", step, "\nCost: ", cost_val, "\n")

    # Accuracy report
    h, c, a = sess.run([y_hat, predicted, accuracy], feed_dict = {X: x_data, Y: y_data})
    print("\ny_hat: ", h, "\nPrediction: ", c, "\nAccuracy: ", a)
```

```python
import pandas as pd
import random
from sklearn import model_selection
```

```python
z = pd.read_csv('EastWestAirlinesNN.csv')
array = z.values

X = array[:, 1:15]
Y = array[:, 15]

test_size = 0.4
seed = 15
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size = test_size, random_state = seed)
```

```python
# h1 5, h2 7, h3 3
X = tf.placeholder(tf.float32, shape = [None, 14])
Y = tf.placeholder(tf.float32)

W1 = tf.Variable(tf.random_normal([14,5]), name = 'weight1')
b1 = tf.Variable(tf.random_normal([5]), name = 'bias1')
M1 = tf.sigmoid(tf.matmul(X, W1) + b1)

W2 = tf.Variable(tf.random_normal([5,7]), name = 'weight2')
b2 = tf.Variable(tf.random_normal([7]), name = 'bias2')
M2 = tf.sigmoid(tf.matmul(M1, W2) + b2)

W3 = tf.Variable(tf.random_normal([7,3]), name = 'weight3')
b3 = tf.Variable(tf.random_normal([3]), name = 'bias3')
M3 = tf.sigmoid(tf.matmul(M2, W3) + b3)

W4 = tf.Variable(tf.random_normal([3,1]), name = 'weight4')
b4 = tf.Variable(tf.random_normal([1]), name = 'bias4')
y_hat = tf.sigmoid(tf.matmul(M3, W4) + b4)
```

```python
# cost function
cost = -tf.reduce_mean(Y * tf.log(y_hat) + (1- Y) * tf.log(1 - y_hat))
train = tf.train.GradientDescentOptimizer(learning_rate = 1e-4).minimize(cost)

# Accuracy
predicted = tf.cast(y_hat > 0.5, dtype = tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype = tf.float32))

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(10001):
    _, cost_val = sess.run([train, cost], feed_dict = {X: X_train, Y: Y_train})
    if step % 100 == 0:
        print("Step: ", step, "\nCost: ", cost_val, "\n")

# Accuracy report
h, c, a = sess.run([y_hat, predicted, accuracy], feed_dict = {X: X_test, Y: Y_test})
print("\ny_hat: ", h, "\nPrediction: ", c, "\nAccuracy: ", a)
```
